```python
import pandas as pd
import numpy as np
```

### 1

```python
stock1 = pd.read_csv('stocks1.csv')
#cho biết kiểu dữ liệu(dtype)của các cột của stock1
print('cho biết kiểu dữ liệu(dtype)của các cột của stock1\n')
print(stock1.dtypes)
#xem thông tin của stock1
print('xem thông tin của stock1\n')
print(stock1.info())
```

```python
stock2 = pd.read_csv('stocks2.csv')
#cho biết kiểu dữ liệu(dtype)của các cột của stock2
print('cho biết kiểu dữ liệu(dtype)của các cột của stock2\n')
print(stock2.dtypes)
#xem thông tin của stock2
print('xem thông tin của stock2\n')
```

```python
companies = pd.read_csv('companies.csv')
#cho biết kiểu dữ liệu(dtype)của các cột của companies
print('cho biết kiểu dữ liệu(dtype)của các cột của companies\n')
print(companies.dtypes)
#xem thông tin của companies
print('xem thông tin của companies\n')
print(companies.info())
```

### 2

```python
stock1['high'] = stock1[["high"]].fillna(stock1.max())
stock1
```

```python
stock1['low'] = stock1[["low"]].fillna(stock1.min())
stock1
```

### 3

```python
stocks = pd.concat([stock1, stock2],axis=  0)
stocks
```

```python
stocks.tail(15)
```

### 4

```python
stocks_companies = pd.concat([stocks, companies], axis = 0)
```

```python
stocks_companies.head()
```

### 5

```python
stocks.groupby('symbol').agg({'open': 'mean','high':'mean','close':'mean','low':'mean','volume':'mean'})
```

### 6

```python
print('giá đóng cửa trung bình,lớn nhất và nhỏ nhất ở mỗi công ty: ')
stocks.groupby('symbol').agg({'close':['mean','max','min']})
```
